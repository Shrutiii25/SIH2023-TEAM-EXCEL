```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
```

```python
df['Target'] = lbl.fit_transform(df['Target'])
```

```python
correlational_matrix = df.corr()
correlational_matrix
```

```python
target_correlation = correlational_matrix['Target']
```

```python
selectedcols = target_correlation[target_correlation > 0.1].index.tolist()
```

```python
df = df[selectedcols]
```

```python
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.boxplot(df)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
mask = ~((df < Q1 - 1.5 * IQR) | (df > Q3 + 1.5 * IQR))
df = df[mask]
```

```python
sns.boxplot(df)
```

```python
df.isna().sum()
```

```python
df = df.fillna(df.median())
```
